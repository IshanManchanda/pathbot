```python
def post(qurl, qanswer) :

    head={'Content-Type': 'application/json'}
    
    answer = qanswer.encode('utf8')
    
    try :
        req = urllib.request.Request(qurl, data=answer, headers=head)
        res = urllib.request.urlopen(req)
        response = json.load(res)

        return response
    
    except urllib.error.HTTPError as e:
        response = json.load(e)
        return response
```

## Helper Functions

```python
def rel_coord(nsew, coordinate) : 
    
    #get the passed in coordinate
    coord = (coordinate[0], coordinate[1])
    
    #translate the passed in move direction, nsew, to coordinates
    if nsew == 'N' : coord = (coordinate[0], coordinate[1] + 1) 
    if nsew == 'S' : coord = (coordinate[0], coordinate[1] - 1)
    if nsew == 'E' : coord = (coordinate[0] + 1, coordinate[1])
    if nsew == 'W' : coord = (coordinate[0] - 1, coordinate[1])
    
    #return the new coordinate (after the move in nsew direction)
    return coord
        
def move_in_dir(pos, nsew, map) :
    
    #given a position, pos, move direction, nsew, and the associated response JSON
    
    #prepare requested move direction for POST function
    mvdir = '{"direction": "' + nsew + '"}'
    
    #return array of new coordinate, after the move and Pathbot response
    return [rel_coord(nsew, pos) , post(baseurl + map['locationPath'], mvdir)]
```

# Breadth-First Search

```python
def search(responses, iterations = -1) :
    
    #initial position
    position = (0,0) 

    #list of visited position
    visited = []
    
    #iteration counter
    i = 0
    
    while responses[0][1]['status'] == 'in-progress' :

        #search logic <^v><^v><^v><^v><^v><^v><^v><^v><^v><^v><^v><^v><^v>
        
        #foreach direction in exits    
        for dirctn in responses[0][1]['exits'] :

            #get pathbot response for that move
            moveto = move_in_dir(position, dirctn, responses[0][1])

            #if not in visited, 
            if moveto[0] not in visited :

                #add it to responses
                responses.append(moveto)

                #add it to visited
                visited.append(moveto[0])

                #if the exit is found *************************************
                if moveto[1]['status'] == 'finished' : return moveto[1]['description']

        #remove top position ------------------------
        responses.pop(0)    
        

        #update position ^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
        position = responses[0][0]
        #position = [x[0] for x in responses][0]
        
        #/search logic <^v><^v><^v><^v><^v><^v><^v><^v><^v><^v><^v><^v><^v>

        #feedback >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
        for j in responses :
            print (i, j[0], j[1]['description'])

        #additional stopping criteria [][][][][][][][][][][][][][][][][][][][][]
        if iterations > 0 :
            if i >= iterations : 
                return 'reached iteration max'                
        
        #iteration incrament
        i += 1            
```
